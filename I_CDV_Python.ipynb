12. Utworzyć skrypt z interfejsem tekstowym obliczający n-ty element ciągu Fibonacciego - wykonać zadanie iteracyjnie i rekurencyjnie

```
def fibo_recurv(n): return n if n <= 1 else (fibo_recurv(n-1) + fibo_recurv(n-2))

def fibo_iter(n):
    a,b = 0,1
    for i in range(n): a,b = b,a+b
    return a

n = int(input('Podaj n: '))
print("Fibonacci rekursywnie: {}".format(fibo_recurv(n)))
print("Fibonacci iteratywnie: {}".format(fibo_iter(n)))
```

13. Utworzyć skrypt z interfejsem tekstowym obliczający symbol Newtona n po k. Utworzyć funkcję, która będzie wywoływać inną funkcję

```
from math import factorial
def binomial(n, k):
    return factorial(n) // (factorial(k) * factorial(n - k))

n = float(input('Podaj n: '))
k = float(input('Podaj k: '))
print("Wynik: {}".format(binomial(n,k)))
```

c) funkcję odwrotnie-potęgową a/x^n
Każda z fukcji powinna przyjmować parametry równania, natomiast zwracać powinna dwie listy - x i y, które następnie będzie można wykreślić na wykresie


```
import numpy as np
from matplotlib import pyplot as plt

def funlin(a, b, step=0.01, range=20):
    x = list(np.arange(-range/2,range/2,step))
    f = list(map(lambda y: a*y+b, x))
    return x, f

def funsqr(a, b, c, step=0.01, range=20):
    x = list(np.arange(-range/2,range/2,step))
    f = list(map(lambda y: a*y**2+b*y+c, x))
    return x, f 

def funcinvexp(a, n, step=0.01, range=20):
    x = list(np.arange(-range/2,range/2,step))
    if 0 in x: x.remove(0)
    f = list(map(lambda y: a/(y**n), x))
    return x, f

x, y = funcinvexp(1, 1, .5)
plt.plot(x, y)
x, y = funlin(1, 1, .5)
plt.plot(x, y)
x, y = funsqr(.1, -.1, -2, .5)
plt.plot(x, y)
plt.show()
```

23. Utwórz funkcję, która jako argument będzie przyjmować listę liczb zmiennoprzecinkowych, a jej wynikiem będzie dominanta (moda). Skorzystaj z obiektu Counter i jego metody most_common z pakietu collections

```
def mode(vals):
    return Counter(vals).most_common(1)[0]

shape, scale = 2., 1.
vals = np.around(np.random.gamma(shape, scale, 100),1)
print(vals)
print(mode(vals))
```

30. Wynegeruj listę 1000 liczb losowych o rozkładzie normalnym. Wykreśl histogram oraz średnią, medianę, dominantę, odchylenie standardowe, wariancję, skośność i kurtozę

```
def fancyhist(data):
  plt.hist(data)
  plt.axvline(np.mean(data), color='r', label='średnia = {}'.format(round(np.mean(data),3)))
  plt.axvline(np.median(data), color='g', label='mediana = {}'.format(round(np.median(data),3)))
  plt.axvline(np.std(data), color='b', label='odchylenie stnd = {}'.format(round(np.std(data),3)))
  plt.axvline(skew(data), color='y', label='skośność = {}'.format(round(skew(data),3)))
  plt.axvline(kurtosis(data), color='black', label='kurtoza = {}'.format(round(kurtosis(data),3)))
  plt.legend(bbox_to_anchor=(1.1, 1), loc='upper left', borderaxespad=0.)
  plt.show()

data = np.random.normal(scale=10, size=1000)
fancyhist(data)
```

34. Utwórz klasę Kwadrat z konstruktorem ustalającym jego bok oraz metodami: wyświetlającymi wartość tego boku, obliczającymi pole i obwód figury

```
class Square():
    def __init__(self, a=0):
        self._flag = False 
        self._a = a
        self._flag = True

    def __str__(self):
        return f'Kwadrat o boku = {self._a}'
    
    def area(self):
        if self._flag:
            return self._a ** 2

    def perimeter(self):
        if self._flag:
            return 4 * self._a

obj = Square(5)
print(obj)
print("Pole kwadradu: {}, obwód kwadradu: {}".format(obj.area(), obj.perimeter()))        
```
